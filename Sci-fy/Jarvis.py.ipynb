```python
engine = pyttsx3.init('sapi5')
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[0].id)
```

```python
def speak(text):
    engine.say(text)
    engine.runAndWait()
```

```python
def wishMe():
    hour = datetime.datetime.now().hour
    if hour >= 0 and hour < 12:
        speak("hello, good morning Mr. Rakshan")
        print("hello, good morning Mr. Rakshan")
    elif hour >= 12 and hour <= 18:
        speak("hello, Good afternoon Mr. Rakshan")
        print("hello, Good afternoon Mr. Rakshan")
    else:
        speak("Its already the night time better go to sleep")
        print("Its already the night time better go to sleep")
```

```python
def takeCommand():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("I AM LISTENING")
        audio = r.listen(source)

    try:
        statement = r.recognize_google(audio, language='en-in')
        print(f"user said:{statement}\n")

    except Exception as e:
        speak("pardon me , please say that again")
        return "None"
    return statement
```

```python
speak("LOADING YOUR PERSONAL A.I ASSISTANT JARVIS")
wishMe()
```
